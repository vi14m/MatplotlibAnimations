```python
import matplotlib.pyplot as plt
import random
import matplotlib.animation as animation
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Linear Regression Animation
x_values = []
y_values = []

reg = LinearRegression()

for i in range(100):
  plt.clf()
  x_values.append(random.randint(0,100))
  y_values.append(random.randint(0,100))

  x=np.array(x_values).reshape(-1,1)
  y=np.array(y_values).reshape(-1,1)

  if i%5==0:
    reg.fit(x,y)
    plt.scatter(x_values,y_values,color='blue')
    plt.plot(list(range(100)),reg.predict(np.array([x for x in range(100)]).reshape(-1,1)))
    plt.pause(0.0001)

plt.show()
```

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                yield arr

# update the bar heights for the animation
def update_bars(arr, bars):
    for bar, val in zip(bars, arr):
        bar.set_height(val)

# create the animation
def animate_sorting(arr):
    fig, ax = plt.subplots()
    bars = ax.bar(range(len(arr)), arr, align='edge')

    ax.set_xlim(0, len(arr))
    ax.set_ylim(0, int(1.1 * max(arr)))

    anim = animation.FuncAnimation(fig, update_bars, frames=bubble_sort(arr), fargs=(bars,), repeat=False)
    plt.show()

# Example usage
np.random.seed(42)
arr = np.random.randint(0, 100, size=30)
animate_sorting(arr)
```

```python
def bubble_sort(arr):
    n = len(arr)
    swapped = True
    for i in range(n):
        if not swapped:
            break
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
            yield arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        yield arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        yield arr

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        yield from quick_sort(arr, low, pi - 1)
        yield from quick_sort(arr, pi + 1, high)
        yield arr

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def update_fig(arr, rects):
    for rect, val in zip(rects, arr):
        rect.set_height(val)

def animate_and_save_sort(sort_func, arr, filename):
    fig, ax = plt.subplots()
    rects = ax.bar(range(len(arr)), arr, align="edge")
    ax.set_xlim(0, len(arr))
    ax.set_ylim(0, int(1.1 * max(arr)))

    anim = animation.FuncAnimation(fig, update_fig, fargs=(rects,), frames=sort_func(arr), repeat=False)
    anim.save(filename, writer='ffmpeg', fps=10)

    plt.show()

N = 50
arr = np.random.randint(1, 100, N)
animate_and_save_sort(bubble_sort, arr.copy(), 'bubble_sort_animation.mp4')
animate_and_save_sort(insertion_sort, arr.copy(), 'insertion_sort_animation.mp4')
animate_and_save_sort(selection_sort, arr.copy(), 'selection_sort_animation.mp4')
animate_and_save_sort(lambda arr: quick_sort(arr, 0, len(arr) - 1), arr.copy(), 'quick_sort_animation.mp4')
```
